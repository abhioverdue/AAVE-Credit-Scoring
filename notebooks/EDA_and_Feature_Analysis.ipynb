```python
import pandas as pd

df = pd.read_json('../data/raw/user-transactions.json')
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.countplot(data=df, x='action', order=df['action'].value_counts().index)
plt.title('Transaction Action Distribution')
plt.show()
```

```python
df['assetSymbol'] = df['actionData'].apply(lambda x: x.get('assetSymbol') if isinstance(x, dict) else None)
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='assetSymbol', order=df['assetSymbol'].value_counts().head(10).index)
plt.xticks(rotation=45)
plt.title('Top 10 Assets')
plt.show()
```

```python
df['amount'] = pd.to_numeric(df['actionData.amount'], errors='coerce')
sns.boxplot(x='action', y='amount', data=df)
plt.yscale('log')
plt.title('Transaction Amounts by Action')
plt.show()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
df.set_index('timestamp').resample('W').size().plot()
plt.title('Weekly Transaction Volume')
plt.show()
```

```python
wallet_counts = df['userWallet'].value_counts()
sns.histplot(wallet_counts, bins=50, log_scale=(False, True))
plt.title('Transaction Count per Wallet')
plt.show()
```
